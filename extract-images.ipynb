```python
# Importing all necessary libraries 
import cv2 
import os 

# Read the video from specified path 
cam = cv2.VideoCapture("/home/nueng/IDS-Videos/04-After-ESRGAN/WW1-4-Outputs-esrgan/WW1-4-3-15-pm-ge_upscaled_3840_2160.mp4") 

try: 

    # creating a folder named data 
	if not os.path.exists('data'): 
		os.makedirs('data') 

# if not created then raise error 
except OSError: 
	print ('Error: Creating directory of data') 

# frame 
currentframe = 0

while(True): 
	
	# reading from frame 
	ret,frame = cam.read() 

	if ret: 
		# if video is still left continue creating images 
		name = '/home/nueng/IDS-Videos/08-extrace-images/04-frame-After-ESRGAN/WW1-4/afternoon/4K/frame' + str(currentframe) + '.jpg' #./data/frame
		print ('Creating...' + name) 

		# writing the extracted images 
		cv2.imwrite(name, frame) 

		# increasing counter so that it will 
		# show how many frames are created 
		currentframe += 1
	else: 
		break

# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows() 
```
